```python
print(DFCountryRank)
```

```python
DFCountryRank.columns.to_list()
```

```python
##FORMATTING##
```

```python
columns=['CountryRegion','CountryName','BirthRateValue','BirthRateRank','EnergyConsumptionValue','EnergyConsumptionRank','CDEmissionsValue','CDEmissionsRank','FertilityRateValue','FertilityRateRank','IPGrowthRateValue','IPGrowthRateRank','LaborForceValue','LaborForceRank','LifeExpectancyValue','LifeExpectancyRank','UnemploymentRateValue','UnemploymentRateRank']
```

```python
DFCountryRank=DFCountryRank[columns]
```

```python
DFCountryRank
```

```python
DFCountryRank.info()
```

```python
from pandas.api.types import CategoricalDtype
```

```python
severalToNum=lambda x:pd.to_numeric(x,errors='coerce')

where=DFCountryRank.columns[2:]

DFCountryRank.loc[:,where]=DFCountryRank.loc[:,where].apply(severalToNum)
```
